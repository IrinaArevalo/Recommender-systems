```python
import numpy as np
import pandas as pd

ratings_data = pd.read_csv("ml-latest-small/ratings.csv") #movielens data 
ratings_data.head()
```

```python
movie_names = pd.read_csv("ml-latest-small/movies.csv")
movie_names.head()
```

```python
movie_data = pd.merge(ratings_data, movie_names, on='movieId')
```

```python
movie_data.head()
```

```python
#average rating of each movie
movie_data.groupby('title')['rating'].mean().head()
```

```python
#order by average rating
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
#ratings by movie (ordered)
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
#movie with average rating and number of ratings
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
ratings_mean_count.head()
```

```python
#histogram of number of ratings
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,6))
_ = ratings_mean_count['rating_counts'].hist(bins=50)
```

```python
#histogram of mean rating
plt.figure(figsize=(8,6))
_ = ratings_mean_count['rating'].hist(bins=50)
```

```python
#table of userId vs movie
user_movie_rating = movie_data.pivot_table(index='userId', columns='title', values='rating')
user_movie_rating.head()
```

```python
forrest_gump_ratings = user_movie_rating['Forrest Gump (1994)']
forrest_gump_ratings.head()
```

```python
#number of ratings by each user
ratings_data.groupby('userId')['movieId'].count().head()
```

```python
#correlation between forrest gump and other movies
movies_like_forest_gump = user_movie_rating.corrwith(forrest_gump_ratings)

corr_forrest_gump = pd.DataFrame(movies_like_forest_gump, columns=['Correlation'])
corr_forrest_gump.dropna(inplace=True)
corr_forrest_gump.head()
```

```python
corr_forrest_gump.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_forrest_gump.sort_values('Correlation', ascending=True).head(10)
```

```python
corr_forrest_gump = corr_forrest_gump.join(ratings_mean_count['rating_counts'])
corr_forrest_gump.head()
```

```python
corr_forrest_gump[corr_forrest_gump ['rating_counts']>50].sort_values('Correlation', ascending=False).head()
```

```python
corr_forrest_gump[corr_forrest_gump ['rating_counts']>50].sort_values('Correlation', ascending=True).head()
```
